```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/drive/MyDrive/adaptive-learning/data/raw/StudentsPerformance.csv')
print('Shape:', df.shape)
print(df.dtypes)
print(df.head(10))
print(df.describe())
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for i, col in enumerate(['math score', 'reading score', 'writing score']):
    axes[i].hist(df[col], bins=20, color='steelblue', edgecolor='white')
    axes[i].set_title(col)
plt.tight_layout()
plt.savefig('/content/drive/MyDrive/adaptive-learning/data/score_distributions.png')
plt.show()
```

```python
categorical_cols = ['gender', 'race/ethnicity', 'parental level of education',
                    'lunch', 'test preparation course']
fig, axes = plt.subplots(1, 5, figsize=(20, 4))
for i, col in enumerate(categorical_cols):
    df[col].value_counts().plot(kind='bar', ax=axes[i], color='steelblue')
    axes[i].set_title(col, fontsize=9)
    axes[i].tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.savefig('/content/drive/MyDrive/adaptive-learning/data/categorical_distributions.png')
plt.show()
```
